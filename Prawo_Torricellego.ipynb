### 1. Wstęp

Dynamika płynów to jeden z wielu fascynujących działów mechaniki klasycznej, w naszym projekcie skupimy się równaniu Torricellego, które przez wielu jest uważane za jedno w największych jej osiągnięć.

### Evangelista Torricelli

Urodził się w 15 października 1608 roku w Faenzie, Włochy. Poświęcił całe życie matematyce, miał ogromny wkład w rozwój rachunku różniczkowego na długo przed jego formalnym sformuowaniem przez Newtona i Leibniza. Stworzył koncept pochodnej wywodzący się z dynamiki, wykazał że styczna w dowolnym punkcie krzywej tworzy z osią czasu kąt, którego tangens mierzy prędkość tego poruszającego się obiektu w tym punkcie. Najważniejszym wkładem Torricellego w rozwój analizy było dostrzeżenie odwrotnej natury różniczkowania i całkowania. Jego notatki były fundamentem do prac nad rachunkiem różniczkowym. Prowadził zaawansowane badania nad maksimami i minimami, rozwiązując m.in. problem punktu o najmniejszej sumie odległości od wierzchołków trójkąta (tzw. punkt Torricellego). Wykazał, że bryła o nieskończonej długości (powstała przez obrót hiperboli wokół jej asymptoty) może mieć skończoną objętość. Bryłę nazwano rogiem Gabriela lub trąbką Torricellego. W 1644 roku przeprowadził słynny eksperyment z rurką wypełnioną rtęcią, który nie tylko doprowadził do wynalezienia barometru, ale udowodnił istnienie próżni (tzw. próżnia Torricellego). Obalił tym samym obowiązujący od wieków dogmat Arystotelesa *horror vavui* (*natura boi się próżni*). Zmarł 25 października 1647 we Florencji<a id="cyt1"></a>[[1]](#bib1).

### Równanie Torricellego

Zjawisko zostało pierwszy raz opisane w 1644 przez ucznia Galileusza, Evangelistę Torricellego w jego dziele "Opera Geometrica"<sup>[[3]](#bib3)</sup>. <a id="cyt3"></a>Torricelli zauważył, że strumień cieczy skierowany pionowo w górę powinien osiągnąć ten sam poziom, na jakim znajduje się lustro wody w zbiorniku. Co oznacza, że prędkość wypływającej cieczy $v$ jest równa prędkości końcowej ciała spadającego swobodnie z wysokości napełnienia $h$<sup>[[4]](#bib4)</sup>.<a id="cyt4"></a>

Załóżmy, że ciecz jest idealna. Z zasady zachowania energii otrzymujemy:
$$E_{\text{kinetyczna}} = E_{\text{potencjalna}}$$
$$\frac{mv^{2}}{2} = mgh$$
Po uproszczeniu:
$$v=\sqrt{2gh}$$

<br clear="right"/>

Z powyższego wzoru wynika, że prędkość osiągnie największą wartość natychmiast, wraz z rozpoczęciem ruchu (wypływaniem cieczy), co jest sprzeczne z zasadą bezwładności.

Torricelli nie odkrył jednak tej zależności korzytając z zasady zachowania energii, lecz okazało się to być szczególnym przypadkiem równania Bernoulliego.

Isaac Newton sprzeciwił się tezie Torricellego w pierwszym wydaniu *Principiów* z 1687<sup>[1](#przyp1)</sup>.<a id="bprzyp1"></a>Uważał, że strumień cieczy wzniesie się jedynie do połowy wysokości napełnienia nad otworem. W drugim wydaniu z 1713 roku poparł tezę Torricellego, lecz w trzecim wydaniu powrócił do poprzedniego stanowiska. Zarówno Torricelli jak i Newton nie brali pod uwagę wpływu bezwładności cieczy. Zakładano, że woda osiąga docelową prędkość natychmiastowo. Pierwszym, który przyjrzał się temu problemowi był Daniel Bernoulli. Powiązał zmiany ciśnienia i wysokości z prędkością. W 1738 sfromuował równanie Bernoulliego:

Załóżmy, że płyn jest nieściśliwy, nie jest lepki oraz przepływ jest stacjonarny i bezwirowy. Wtedy:

$$e_{m} = \frac{v^{2}}{2}+gh+\frac{p}{\rho} = const,$$
gdzie:

$e_{m}$ – energia jednostki masy płynu,

$\rho$ – gęstość płynu,

$v$ – prędkość płynu w rozpatrywanym miejscu,

$h$ – wysokość w układzie odniesienia, w którym liczona jest energia potencjalna,

$g$ – przyspieszenie grawitacyjne,

$p$– ciśnienie płynu w rozpatrywanym miejscu.


Zastosujmy równanie Bernoulliego dla dwóch punktów: jednego na powierzchni cieczy ($z_1 = h$), a drugiego w otworze na wysokości $z_2$. Przymijmy, że $z_2 = 0$ (Znajduje się na dnie zbiornika). Wtedy:

$$\frac{v_1^{2}}{2}+gz_1+\frac{p_1}{\rho} = \frac{v_2^{2}}{2}+gz_2+\frac{p_2}{\rho}$$

Załóżmy, że prędkość na powierzchni cieczy równa sie zero oraz zauważmy że $p_1 = p_2 = p_{atmosferyczne}$
$$ gh = \frac{v_2^{2}}{2}$$
$$ v_2 = \sqrt{2gh} $$
Zatem, wzór Torricellego jest szczególnym przypadkiem równania Bernoulliego.

### 2. Opis zjawiska/podstawy teoretyczne

Prędkość obliczona za pomocą wzoru Torricellego jest prędkością teoretyczną, w rzeczywistości prędkość wypływu jest niższa. Wpływają na to:
* pominięcie lepkości cieczy, tarcie wewnętrzne cząsteczek cieczy powoduje straty energii
* Kontrakcja strumienia (Vena contracta): Strumień cieczy po przejściu przez otwór ulega zwężeniu, co sprawia, że efektywne pole powierzchni wypływu jest mniejsze niż pole otworu $a$<sup>[[6]](#bib6)</sup>.<a id="cyt6"></a>

Aby uwzględnić te straty, często stosuje się współczynnik $\mu$ taki, że $$v_{real} = \mu \sqrt{2gh},$$ gdzie dla wody $\mu \approx 0.6$ do $0.65$.

Rozważmy jednak model idealny $v = \sqrt{2gh}$.
Niech:

$A = A(t)$ - pole powierzchni przekroju poprzecznego zbiornika (lustra wody),

$a = const$ - pole powierzchni otworu wylotowego,

$h = h(t)$ - wysokość słupa cieczy,

$v = v(t)$ - prędkość wypływu cieczy.

Zmiana objętości cieczy w zbiorniku musi się równać objętości cieczy, która wypłynęła przez otwór. Zatem:
$$ -A \cdot dh = a \cdot v \cdot dt $$

Otrzymujemy równanie różniczkowe pierwszego rzędu opisujące prędkość zmiany wysokości:
$$\frac{dh}{dt} = -\frac{a}{A}v \tag{I}$$
<a id="ad1"></a>

Przekształcając wzór Torricellego otrzymujemy:

$$ h = \frac{v^2}{2g} $$

Różniczkujemy obie strony względem zmiennej $t$:
$$\frac{dh}{dt} = \frac{d}{dt}(\frac{v^{2}}{2g})$$
$$\frac{dh}{dt} = \frac{2v}{2g}\frac{dv}{dt} = \frac{v}{g}\frac{dv}{dt}\tag{II}$$

Podstawmy do (II) do równania (I):

$$ \frac{dh}{dt} = -\frac{a}{A}v $$
$$ \frac{v}{g}\frac{dv}{dt} = -\frac{a}{A}v $$

Ostatecznie, otrzymujemy równanie różniczkowe opisujące zmianę prędkości przepływu cieczy:
$$ \frac{dv}{dt} = -\frac{ag}{A} $$

* Rozważmy przypadek $A=A(t)=const$, jest to przypadek kiedy przekrój poprzeczny nie zmienia się, naczynie jest np walcem, sześcianem.
$$ \frac{dv}{dt} = -\frac{ag}{A} $$
Jest to równanie o zmiennych rozdzielonych. Całkujemy obie strony względem czasu:
$$\int_{v_0}^{v(t)} dv = \int_{0}^{t} -\frac{ag}{A} dt$$
$$v(t) - v_0 = -\frac{ag}{A} t$$
$$v(t) = v_0 - \frac{ag}{A}t$$

Prędkość wypływu maleje liniowo w czasie. Wykresem jest linia prosta spadająca w dół, aż do osiągnięcia $v=0$.

* Rozważmy przypadek A = A(t) będzie powierzchnią przekroju poprzecznego stożka o promieniu górnym $R$ i wysokości $H$. 

Promień lustra wody $r$ na wysokości $h$ wynosi: 
$$r=r(h) = \frac{R}{H}h.$$

Zatem pole powierzchni lustra cieczy $A$ wynosi:
$$A = A(h) = \pi \cdot \left(r(h)\right)^2 = \frac{\pi R^2}{H^2} \cdot h^2$$

Wykorzystując związek $h = \frac{v^2}{2g}$, podstawiamy go do wzoru na pole:
$$A(v) = \frac{\pi R^2}{H^2} \cdot \left( \frac{v^2}{2g} \right)^2 = \frac{\pi R^2}{4g^2 H^2} \cdot v^4$$

Podstawiając $A(v)$ do naszego głównego równania $\frac{dv}{dt} = -\frac{ag}{A}$, otrzymujemy:
$$\frac{dv}{dt} = -\frac{ag}{\frac{\pi R^2}{4g^2 H^2} \cdot v^4} = -\frac{4ag^3 H^2}{\pi R^2} \cdot \frac{1}{v^4}$$

Niech $-\frac{4ag^3 H^2}{\pi R^2} = K$, wtedy:
$$\frac{dv}{dt} = K \cdot \frac{1}{v^4}$$
$$\int_{v_0}^{v(t)} v^{4}dv = \int_{0}^{t} -\frac{ag}{A} Kdt$$
$$\frac{1}{5}\left(v(t)^{5}-v_0^{5}\right) = Kt$$
$$v(t) = \sqrt[5]{v_0^{5}+5Kt}$$
$$v(t) = \sqrt[5]{v_0^{5}-\frac{20ag^3 H^2}{\pi R^2} t}$$

W stożku prędkość początkowo maleje bardzo powoli (szeroka góra). Gdy poziom wody zbliża się do dna (wąska część), prędkość $v(t)$ gwałtownie spada do zera.

* W przypadku uwzględnienia współczynnika $\mu$ równanie różniczkowe przyjmuje postać:
$$v = \mu \sqrt{2gh}$$
$$h=\frac{v^{2}}{2\mu g}$$
$$\frac{dh}{dt} = \frac{dh}{dt} \left(\frac{v^{2}}{2\mu g}\right) = \frac{v}{\mu g}\frac{dv}{dt} \tag{III}$$

Korzystamy z zasady zachowania masy:
$$ -A \cdot dh = a \cdot v \cdot dt $$
$$ \frac{dh}{dt} = -\frac{a}{A}v $$
Wstawiamy (III):
$$\frac{v}{\mu g}\frac{dv}{dt} = -\frac{a}{A}v$$
$$\frac{dv}{dt} = -\frac{a \cdot \mu \cdot g}{A}$$


Końcowa postać $v=v(t)$ wygląda odpowienio:

$v(t) = v_0 - \left( \frac{a \mu g}{A} \right) t$, dla A = const 

$v(t) = \sqrt[5]{v_0^5 - \left(\frac{20 a \mu g^3 H^2}{\pi R^2} \right)t}$ dla stożka.
