### 1. Wstęp

Celem niniejszego projektu jest analiza dynamiki wypływu cieczy przez mały otwór, wyprowadzenie teoretycznego modelu matematycznego oraz weryfikacja i symulacja czasu opróżniania zbiornika w zależności od jego geometrii. Modelowanie wypływu cieczy jest kluczowe w projektowaniu systemów irygacyjnych, zbiorników paliwowych oraz w przemyśle chemicznym. Zrozumienie, jak szybko opróżnia się zbiornik, pozwala na optymalizację procesów technologicznych i zapewnienie bezpieczeństwa. 

### Evangelista Torricelli

Urodził się w 15 października 1608 roku w Faenzie, Włochy. Poświęcił całe życie matematyce, miał ogromny wkład w rozwój rachunku różniczkowego na długo przed jego formalnym sformuowaniem przez Newtona i Leibniza. Stworzył koncept pochodnej wywodzący się z dynamiki, wykazał że styczna w dowolnym punkcie krzywej tworzy z osią czasu kąt, którego tangens mierzy prędkość poruszającego się obiektu w punkcie. Najważniejszym wkładem Torricellego w rozwój analizy było dostrzeżenie odwrotnej natury różniczkowania i całkowania. Jego notatki były fundamentem do prac nad rachunkiem różniczkowym. Prowadził zaawansowane badania nad maksimami i minimami, rozwiązując m.in. problem punktu o najmniejszej sumie odległości od wierzchołków trójkąta (tzw. punkt Torricellego). Wykazał, że bryła o nieskończonej długości (powstała przez obrót hiperboli wokół jej asymptoty) może mieć skończoną objętość. Bryłę nazwano rogiem Gabriela lub trąbką Torricellego. W 1644 roku przeprowadził słynny eksperyment z rurką wypełnioną rtęcią, który nie tylko doprowadził do wynalezienia barometru, ale udowodnił istnienie próżni (tzw. próżnia Torricellego). Obalił tym samym obowiązujący od wieków dogmat Arystotelesa *horror vavui* (*natura boi się próżni*). Zmarł 25 października 1647 we Florencji<a id="cyt1"></a><sup>[[1]](#bib1)</sup>.

Zjawisko zostało pierwszy raz opisane w 1644 przez ucznia Galileusza, Evangelistę Torricellego w jego dziele "Opera Geometrica"<sup>[[3]](#bib3)</sup>. <a id="cyt3"></a>Torricelli zauważył, że strumień cieczy skierowany pionowo w górę powinien osiągnąć ten sam poziom, na jakim znajduje się lustro wody w zbiorniku. Co oznacza, że prędkość wypływającej cieczy $v$ jest równa prędkości końcowej ciała spadającego swobodnie z wysokości napełnienia $h$<sup>[[4]](#bib4)</sup>.<a id="cyt4"></a>

Zasada Torricellego opiera się na analogii między ruchem cieczy a kinematyką punktu materialnego. Ciśnienie hydrostatyczne panujące na głębokości $h$ wykonuje nad cieczą taką samą pracę, jaką wykonałaby siła ciężkości nad ciałem spadającym swobodnie z tej samej wysokości. W efekcie energia potencjalna słupa cieczy zostaje w całości zamieniona na energię kinetyczną strumienia wypływowego.

Załóżmy, że ciecz jest idealna. Z zasady zachowania energii otrzymujemy:
$$E_{\text{kinetyczna}} = E_{\text{potencjalna}}$$
$$\frac{mv^{2}}{2} = mgh$$
Po uproszczeniu:
$$v=\sqrt{2gh}$$

<br clear="right"/>

Z powyższego wzoru wynika, że prędkość osiągnie największą wartość natychmiast, wraz z rozpoczęciem ruchu (wypływaniem cieczy), co jest sprzeczne z zasadą bezwładności.

Torricelli nie odkrył jednak tej zależności korzytając z zasady zachowania energii, lecz okazało się to być szczególnym przypadkiem równania Bernoulliego.

Isaac Newton sprzeciwił się tezie Torricellego w pierwszym wydaniu *Principiów* z 1687<sup>[1](#przyp1)</sup>.<a id="bprzyp1"></a>Uważał, że strumień cieczy wzniesie się jedynie do połowy wysokości napełnienia nad otworem. W drugim wydaniu z 1713 roku poparł tezę Torricellego, lecz w trzecim wydaniu powrócił do poprzedniego stanowiska. Zarówno Torricelli jak i Newton nie brali pod uwagę wpływu bezwładności cieczy. Zakładano, że woda osiąga docelową prędkość natychmiastowo. Pierwszym, który przyjrzał się temu problemowi był Daniel Bernoulli. Powiązał zmiany ciśnienia i wysokości z prędkością. W 1738 sfromuował równanie Bernoulliego:

Załóżmy, że płyn jest nieściśliwy, nie jest lepki oraz przepływ jest stacjonarny i bezwirowy. Wtedy:

$$e_{m} = \frac{v^{2}}{2}+gh+\frac{p}{\rho} = const,$$
gdzie:

$e_{m}$ – energia jednostki masy płynu,

$\rho$ – gęstość płynu,

$v$ – prędkość płynu w rozpatrywanym miejscu,

$h$ – wysokość w układzie odniesienia, w którym liczona jest energia potencjalna,

$g$ – przyspieszenie grawitacyjne,

$p$ – ciśnienie płynu w rozpatrywanym miejscu.


Zastosujmy równanie Bernoulliego dla dwóch punktów: jednego na powierzchni cieczy ($z_1 = h$), a drugiego w otworze na wysokości $z_2$. Przymijmy, że $z_2 = 0$ (Znajduje się na dnie zbiornika). Wtedy:

$$\frac{v_1^{2}}{2}+gz_1+\frac{p_1}{\rho} = \frac{v_2^{2}}{2}+gz_2+\frac{p_2}{\rho}$$

Załóżmy, że prędkość na powierzchni cieczy równa sie zero oraz zauważmy że $p_1 = p_2 = p_{atmosferyczne}$
$$ gh = \frac{v_2^{2}}{2}$$
$$ v_2 = \sqrt{2gh} $$
Zatem, wzór Torricellego jest szczególnym przypadkiem równania Bernoulliego.

### 2. Opis zjawiska/podstawy teoretyczne

Prędkość obliczona za pomocą wzoru Torricellego jest prędkością teoretyczną, w rzeczywistości prędkość wypływu jest niższa. Wpływają na to:
* pominięcie lepkości cieczy, tarcie wewnętrzne cząsteczek cieczy powoduje straty energii
* kontrakcja strumienia (Vena contracta): Strumień cieczy po przejściu przez otwór ulega zwężeniu, co sprawia, że efektywne pole powierzchni wypływu jest mniejsze niż pole otworu $a$<sup>[[6]](#bib6)</sup>.<a id="cyt6"></a>

Aby uwzględnić te straty, często stosuje się współczynnik $\mu$ taki, że $$v_{real} = \mu \sqrt{2gh},$$ gdzie dla wody $\mu \approx 0.6$ do $0.65$.

Rozważmy jednak model idealny $v = \sqrt{2gh}$.
Niech:

$A = A(t)$ - pole powierzchni przekroju poprzecznego zbiornika (lustra wody),

$a = const$ - pole powierzchni otworu wylotowego,

$h = h(t)$ - wysokość słupa cieczy,

$v = v(t)$ - prędkość wypływu cieczy.

Zmiana objętości cieczy w zbiorniku musi się równać objętości cieczy, która wypłynęła przez otwór. Zatem:
$$dV_{zbiornika} = dV_{wypływu}$$
$$ -A \cdot dh = a \cdot v \cdot dt $$

Otrzymujemy równanie różniczkowe pierwszego rzędu opisujące prędkość zmiany wysokości:
$$\frac{dh}{dt} = -\frac{a}{A}v \tag{I}$$
<a id="ad1"></a>

Przekształcając wzór Torricellego otrzymujemy:

$$ h = \frac{v^2}{2g} $$

Różniczkujemy obie strony względem zmiennej $t$:
$$\frac{dh}{dt} = \frac{d}{dt}(\frac{v^{2}}{2g})$$
$$\frac{dh}{dt} = \frac{2v}{2g}\frac{dv}{dt} = \frac{v}{g}\frac{dv}{dt}\tag{II}$$

Podstawmy do (II) do równania (I):

$$ \frac{dh}{dt} = -\frac{a}{A}v $$
$$ \frac{v}{g}\frac{dv}{dt} = -\frac{a}{A}v $$

Ostatecznie, otrzymujemy równanie różniczkowe opisujące zmianę prędkości przepływu cieczy:
$$ \frac{dv}{dt} = -\frac{ag}{A} $$

Zauważmy, że zmiana prędkości wypływu cieczy jest proporcjonalna do $-\frac{ag}{A}$, czyli im mniejszy otwór w stosunku do powierzchni lustra cieczy, tym wolniej spada prędkość. Znak minus w równaniu $\frac{dv}{dt} = -\frac{ag}{A}$ ma fundamentalne znaczenie fizyczne - informuje nas o sprzężeniu zwrotnym ujemnym w układzie: Wypływ cieczy zmniejsza wysokość słupa wody ($h$). Zmniejszenie $h$ redukuje energię potencjalną, co bezpośrednio przekłada się na spadek prędkości $v$. Układ dąży do stanu równowagi, w którym $v=0$. 

W układach teoretycznych pole otworu $a$ zazwyczaj przyjmujemy jako stałe. Warto jednak zauważyć, że jest to jedyny parametr (poza zmianą kształtu naczynia), który pozwala na aktywne sterowanie procesem. Stopniowo zamykając otwór (zmniejszając $a$) będziemy spowalniali proces, wartość $\frac{dv}{dt}$ będzie dążyć do zera zatem możemy wydłużyć czas opróżniania zbiornika niemal w nieskończoność, utrzymując bardzo małą, ale stabilną prędkość wypływu. Zwiększając $a$ przyśpieszymy proces. Zmiana $a$ w rzeczywistości realizowana jest poprzez mechaniczne ograniczanie światła otworu lub otwieranie kolejnych śluz. Ma to zastosowanie w przypadku awarii lub w próbie kontrolowania przepływu wody.

* Rozważmy przypadek $A=A(t)=const$, jest to przypadek kiedy przekrój poprzeczny nie zmienia się, naczynie jest np walcem, sześcianem.
$$ \frac{dv}{dt} = -\frac{ag}{A} $$
Jest to równanie o zmiennych rozdzielonych. Całkujemy obie strony względem czasu:
$$\int_{v_0}^{v(t)} dv = \int_{0}^{t} -\frac{ag}{A} dt$$
$$v(t) - v_0 = -\frac{ag}{A} t$$
$$v(t) = v_0 - \frac{ag}{A}t$$
Gdzie: $v_0=\sqrt{2gh(0)}$

Prędkość wypływu maleje liniowo w czasie. Wykresem jest linia prosta spadająca w dół, aż do osiągnięcia $v=0$.

* Rozważmy przypadek $A = A(t)$, gdzie $A$ będzie powierzchnią przekroju poprzecznego stożka o promieniu górnym $R$ i wysokości $H$. 

Promień lustra wody $r$ na wysokości $h$ wynosi: 
$$r=r(h) = \frac{R}{H}h$$

Zatem pole powierzchni lustra cieczy $A$ wynosi:
$$A = A(h) = \pi \cdot \left(r(h)\right)^2 = \frac{\pi R^2}{H^2} \cdot h^2$$

Wykorzystując związek $h = \frac{v^2}{2g}$, podstawiamy go do wzoru na pole:
$$A(v) = \frac{\pi R^2}{H^2} \cdot \left( \frac{v^2}{2g} \right)^2 = \frac{\pi R^2}{4g^2 H^2} \cdot v^4$$

Podstawiając $A(v)$ do naszego głównego równania $\frac{dv}{dt} = -\frac{ag}{A}$, otrzymujemy:
$$\frac{dv}{dt} = -\frac{ag}{\frac{\pi R^2}{4g^2 H^2} \cdot v^4} = -\frac{4ag^3 H^2}{\pi R^2} \cdot \frac{1}{v^4}$$

Niech $-\frac{4ag^3 H^2}{\pi R^2} = K$, wtedy:
$$\frac{dv}{dt} = K \cdot \frac{1}{v^4}$$
$$\int_{v_0}^{v(t)} v^{4}dv = \int_{0}^{t} -\frac{ag}{A} Kdt$$
$$\frac{1}{5}\left(v(t)^{5}-v_0^{5}\right) = Kt$$
$$v(t) = \sqrt[5]{v_0^{5}+5Kt}$$
$$v(t) = \sqrt[5]{v_0^{5}-\frac{20ag^3 H^2}{\pi R^2} t}$$
Gdzie: $v_0=\sqrt{2gh(0)}=\sqrt{2gH}$


W stożku prędkość początkowo maleje bardzo powoli (szeroka góra). Gdy poziom wody zbliża się do dna (wąska część), prędkość $v(t)$ gwałtownie spada do zera.

* W przypadku uwzględnienia współczynnika $\mu$ końcowa postać $v=v(t)$ wygląda odpowienio:

$v(t) = v_0 - \left( \frac{a \mu g}{A} \right) t$, dla A = const 

$v(t) = \sqrt[5]{v_0^5 - \left(\frac{20 a \mu g^3 H^2}{\pi R^2} \right)t}$, dla stożka.

Wróćmy do zasady zachowania masy, lecz wyprowadzimy wzór na wysokość słupa cieczy od czasu $h(t)$ oraz całkowity czas opróżniania $T$. 
* Rozważmy przypadek $A=A(t)=const$
$$ -A \cdot dh = a \cdot v \cdot dt $$
Podstawmy $v=\sqrt{2gh}$:
$$-A \cdot dh = a \cdot \sqrt{2gh} \cdot dt$$
$$\frac{dh}{\sqrt{h}} = -\frac{a \sqrt{2g}}{A} dt$$
Nakładamy całki na obie strony równania. Całkujemy od stanu początkowego ($t=0, h=H$) do stanu w dowolnej chwili $t$ ($h=h(t)$):
$$\int_{H}^{h(t)} \frac{1}{\sqrt{h}} dh = -\int_{0}^{t} \frac{a \sqrt{2g}}{A} dt$$
Po obliczeniu całek ($\int h^{-1/2} dh = 2\sqrt{h}$):
$$2\sqrt{h(t)} - 2\sqrt{H} = -\frac{a \sqrt{2g}}{A} t$$
$$\sqrt{h(t)} = \sqrt{H} - \frac{a \sqrt{2g}}{2A} t$$
Otrzymujemy wzór na wysokość wody w czasie:
$$h(t) = \left( \sqrt{H} - \frac{a \sqrt{2g}}{2A} t \right)^2$$

Zbiornik będzie pusty, gdy wysokość $h(T) = 0$. Podstawiamy to do naszego wyprowadzonego wzoru:
$$0 = \sqrt{H} - \frac{a \sqrt{2g}}{2A} T$$
Przenosimy wyrazy i wyznaczamy $T$:
$$\frac{a \sqrt{2g}}{2A} T = \sqrt{H}$$
$$T = \frac{2A \sqrt{H}}{a \sqrt{2g}}$$
$$T = \frac{A}{a} \sqrt{\frac{2H}{g}}$$

* Uwzględniając współczynnik $\mu$:
$h(t) = \left( \sqrt{H} - \frac{\mu \cdot a \sqrt{2g}}{2A} t \right)^2$ oraz $T = \frac{A}{\mu \cdot a} \sqrt{\frac{2H}{g}}$

* Wróćmy do przypadku ogólnego, $A$ zmienia się w zależności od wysokośći $h$.
$$ -A \cdot dh = a \cdot v \cdot dt $$
Gdzie: $v = \sqrt{2gh}$, zatem
$$ -A \cdot dh = a \cdot \sqrt{2gh} \cdot dt $$
$$ \frac{A(h)dh}{\sqrt{h}} = -a\sqrt{2g}dt$$
Nakładamy całki na obie strony równania:
$$\int{\frac{A(h)}{\sqrt{h}}} dh = -\int{a \sqrt{2g}}{A} dt = -a \sqrt{2g}\int{}dt$$

Rozważmy zbiornik stożkowy, o promieniu górnym $R$ i wysokości $H$, według wcześniejszych obliczeń:
$$A = A(h) = \pi \cdot \left(r(h)\right)^2 = \frac{\pi R^2}{H^2} \cdot h^2$$
Zatem:
$$\int_{H}^{h(t)}{\frac{\frac{\pi R^2}{H^2} \cdot h^2}{\sqrt{h}}} dh = -a \sqrt{2g}\int_{0}^{t}{}dt$$
$$\int_{H}^{h(t)}\frac{\pi R^2}{H^2}\frac{h^2}{\sqrt{h}} dh = -a \sqrt{2g}\int_{0}^{t}{}dt$$
$$\frac{\pi R^2}{H^2}\int_{H}^{h(t)}{h^{\frac{3}{2}}} dh = -a \sqrt{2g}t$$
$$\frac{\pi R^2}{H^2}\frac{2}{5}\left(h(t)^{\frac{5}{2}}-H^{\frac{5}{2}}\right) = -a \sqrt{2g}t$$
$$\frac{2}{5} \left( h^{5/2} - H^{5/2} \right) = -\frac{a \sqrt{2g} \cdot H^2}{\pi R^2} t$$
$$h^{5/2} = H^{5/2} - \frac{5a \sqrt{2g} \cdot H^2}{2 \pi R^2} t$$
$$h(t) = \left( H^{5/2} - \frac{5a \sqrt{2g} \cdot H^2}{2 \pi R^2} t \right)^{2/5}$$
Wiedząc, że $\pi R^2$ to pole powierzchni góry $A_0$, możemy zapisać:
$$h(t) = \left( H^{5/2} - \frac{5a \sqrt{2g} \cdot H^2}{2A_0} t \right)^{2/5}$$

Zbiornik opróżni się całkowicie, gdy $h(T) = 0$. Podstawiamy to do równania:
$$0 = H^{5/2} - \frac{5a \sqrt{2g} \cdot H^2}{2 \pi R^2} T
$$Przekształcamy, aby wyznaczyć $T$:
$$\frac{5a \sqrt{2g} \cdot H^2}{2 \pi R^2} T = H^{5/2}$$
$$T = \frac{2 \pi R^2 \cdot H^{5/2}}{5a \sqrt{2g} \cdot H^2}$$
Skracamy potęgi: $H^{5/2} / H^2 = H^{1/2} = \sqrt{H}$
$$T = \frac{2 \pi R^2 \sqrt{H}}{5 a \sqrt{2g}}$$
$$T = \frac{2 A_0}{5a} \sqrt{\frac{H}{2g}}$$

* Mając powyższe wzory możemy łatwo wyprowadzić wzory na objętość $V$ cieczy pozostałej w zbiorniku

Dla $A=A(t)=const$:
$$V(t)=A(t) \cdot h(t) = A \cdot \left( \sqrt{H} - \frac{a \sqrt{2g}}{2A} t \right)^2$$
$$V(t) = A \cdot (\sqrt{H})^2 \cdot \left( 1 - \frac{\mu a \sqrt{2g}}{2A\sqrt{H}} t \right)^2$$
Ponieważ $A \cdot H = V_0$ (objętość początkowa), otrzymujemy:
$$V(t) = V_0 \cdot \left( 1 - \frac{a \sqrt{2g}}{2A\sqrt{H}} t \right)^2$$
Wykorzystując, że $T = \frac{A}{a} \sqrt{\frac{2H}{g}}$ otrzymujemy:
$$V(t) = V_0 \cdot \left( 1 - \frac{t}{T} \right)^2$$

Dla stożka:
$$V(t) = \frac{1}{3} \pi r(t)^2 h(t)$$
Wiemy, że $r(t) = \frac{R}{H} h(t)$. Podstawiając to do wzoru na objętość, otrzymujemy:
$$V(t) = \frac{1}{3} \pi \left( \frac{R}{H} h(t) \right)^2 h(t) = \frac{\pi R^2}{3 H^2} h(t)^3$$
Podstawiamy wyliczone wcześniej $h(t)$ do wzoru na objętość:
$$V(t) = \frac{\pi R^2}{3 H^2} \left[ \left( H^{5/2} - \frac{5a \sqrt{2g} \cdot H^2}{2 \pi R^2} t \right)^{2/5} \right]^3$$
$$V(t) = \frac{\pi R^2}{3 H^2} \left( H^{5/2} - \frac{5a \sqrt{2g} \cdot H^2}{2 \pi R^2} t \right)^{6/5}$$
Po przekształceniach otrzymujemy:
$$V(t) = V_0 \cdot \left( 1 - \frac{t}{T} \right)^{6/5}$$
