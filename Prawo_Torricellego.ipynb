### 1. Wstęp

Dynamika płynów to jeden z wielu fascynujących działów mechaniki klasycznej, w naszym projekcie skupimy się równaniu Torricellego, które przez wielu jest uważane za jedno w największych jej osiągnięć.

### Evangelista Torricelli

Urodził się w 15 października 1608 roku w Faenzie, Włochy. Poświęcił całe życie matematyce, miał ogromny wkład w rozwój rachunku różniczkowego na długo przed jego formalnym sformuowaniem przez Newtona i Leibniza. Stworzył koncept pochodnej wywodzący się z dynamiki, wykazał że styczna w dowolnym punkcie krzywej tworzy z osią czasu kąt, którego tangens mierzy prędkość tego poruszającego się obiektu w tym punkcie. Najważniejszym wkładem Torricellego w rozwój analizy było dostrzeżenie odwrotnej natury różniczkowania i całkowania. Jego notatki były fundamentem do prac nad rachunkiem różniczkowym. Prowadził zaawansowane badania nad maksimami i minimami, rozwiązując m.in. problem punktu o najmniejszej sumie odległości od wierzchołków trójkąta (tzw. punkt Torricellego). Wykazał, że bryła o nieskończonej długości (powstała przez obrót hiperboli wokół jej asymptoty) może mieć skończoną objętość. Bryłę nazwano rogiem Gabriela lub trąbką Torricellego. W 1644 roku przeprowadził słynny eksperyment z rurką wypełnioną rtęcią, który nie tylko doprowadził do wynalezienia barometru, ale udowodnił istnienie próżni (tzw. próżnia Torricellego). Obalił tym samym obowiązujący od wieków dogmat Arystotelesa *horror vavui* (*natura boi się próżni*). Zmarł 25 października 1647 we Florencji<sup>[[1]](#bib1)</sup>.

### Równanie Torricellego

Zjawisko zostało pierwszy raz opisane w 1644 przez ucznia Galileusza, Evangelistę Torricellego w jego dziele "Opera Geometrica"<sup>[[3]](#bib3)</sup>. Torricelli zauważył, że strumień cieczy skierowany pionowo w górę powinien osiągnąć ten sam poziom, na jakim znajduje się lustro wody w zbiorniku.

### 2. Opis zjawiska/podstawy teoretyczne

Prawo Torricellego pozwala obliczyć prędkość wypływu cieczy z naczynia przez niewielki otwór.
Opisane przez włoskiego fizyka Evangelistę Torricellego w XVII wieku.
Stwierdza ono że ciecz wypływająca pod wpływem grawitacji osiąga tą samą prędkość jaką osiągneło by ciało spadające z wysokości słupa wody  
Wzór:<br>$$v=\sqrt{2gh} \\
  v \text{\:--\:prędkość wypływu cieczy(}m/s \text{)}\\
  g \text{\:--\:przyspieszenie ziemskie(}m/s^2 \text{)}\\
  h \text{\:--\:wysokość słupa cieczy nad otworem(}m \text{)}$$
Wzór ten działa przy założeniach ,że ciecz jest nielepka i nieściśliwa oraz średnica naczynia jest dużo większa od średnicy otworu przez który ciecz wypływa
