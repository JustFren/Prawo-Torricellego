### 2. Opis zjawiska/podstawy teoretyczne

Prawo Torricellego pozwala obliczyć prędkość wypływu cieczy z naczynia przez niewielki otwór.
Opisane przez włoskiego fizyka Evangelistę Torricellego w XVII wieku.
Stwierdza ono że ciecz wypływająca pod wpływem grawitacji osiąga tą samą prędkość jaką osiągneło by ciało spadające z wysokości słupa wody  
Wzór:<br>$$v=\sqrt{2gh} \\
  v \text{\:--\:prędkość wypływu cieczy(}m/s \text{)}\\
  g \text{\:--\:przyspieszenie ziemskie(}m/s^2 \text{)}\\
  h \text{\:--\:wysokość słupa cieczy nad otworem(}m \text{)}$$
Wzór ten działa przy założeniach ,że ciecz jest nielepka i nieściśliwa oraz średnica naczynia jest dużo większa od średnicy otworu przez który ciecz wypływa
